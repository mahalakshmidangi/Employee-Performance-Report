```python
import pandas as pd
```

```python
df = pd.read_excel('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls')
```

```python
df.isnull().sum()
```

```python
print('No of Duplicated rows =',df.duplicated().sum())
```

#### These features have no impact on PerformanceRating from data_exploratory_analysis. So dropping these features.

```python
df.drop(['EmpNumber','BusinessTravelFrequency','EmpJobInvolvement','EmpRelationshipSatisfaction','TrainingTimesLastYear'],axis=1,inplace=True)
```

#### Some features are not ordinal data, so we have to convert it into numerical value by using one hot encoding method.

```python
Gender=pd.get_dummies(data=df['Gender'])
EducationBackground=pd.get_dummies(data=df['EducationBackground'])
MaritalStatus=pd.get_dummies(data=df['MaritalStatus'])
EmpDepartment=pd.get_dummies(data=df['EmpDepartment'])
EmpJobRole=pd.get_dummies(data=df['EmpJobRole'])
OverTime=pd.get_dummies(data=df['OverTime'])
Attrition=pd.get_dummies(data=df['Attrition'])

df.drop(['Gender','EducationBackground','MaritalStatus','EmpDepartment','EmpJobRole','OverTime','Attrition'],axis=1,inplace=True)

final_df=pd.concat([Gender,EducationBackground,MaritalStatus,EmpDepartment,EmpJobRole,OverTime,Attrition,df],axis=1)
```

```python
print("After one hot ending data size :",final_df.shape)
```
