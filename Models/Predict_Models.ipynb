# Predict Models

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import pickle as pkl
```

```python
final_df = pd.read_excel('Processed_Data.xlsx')
```

```python
pd.set_option('display.max_columns',None)
final_df
```

```python
X=final_df.drop('PerformanceRating',axis=1)
Y=final_df['PerformanceRating']
```

```python
MMS=MinMaxScaler()
X_scaled=MMS.fit_transform(X)
X_scaled_df=pd.DataFrame(data=X_scaled,columns=X.columns)
X_scaled_df
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X_scaled_df, Y, test_size=0.20, random_state=42)
```

```python
print('Shape of Xtrain :',Xtrain.shape)
print('Shape of Xtest :',Xtest.shape)
```

### Loading All Models from pickle file

#### 1) Decision Tree Prediction

```python
DTC = pkl.load(open('DTC.pkl','rb'))
DTC_Ypred = DTC.predict(Xtest)
DTC_Ypred
```

#### 2) Random Forest Prediction

```python
RFC = pkl.load(open('RFC.pkl', 'rb'))
RFC_Ypred = RFC.predict(Xtest)
RFC_Ypred
```

#### 3) Gradient Boosting Prediction

```python
GBC = pkl.load(open('GBC.pkl', 'rb'))
GBC_Ypred = GBC.predict(Xtest)
GBC_Ypred
```

```python
plt.figure(figsize=(8,4))
plt.title('All Model Comparision with Actual value')
sns.lineplot(x=DTC_Ypred,y=Ytest.index,color='red',label='DTC')
sns.lineplot(x=RFC_Ypred,y=Ytest.index,color='green',label='RFC')
sns.lineplot(x=GBC_Ypred,y=Ytest.index,color='blue',label='GBC')
sns.lineplot(x=Ytest,y=Ytest.index,color='yellow',label='Actual')
plt.show()
```
