# Visualize

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# !pip install xlrd
```

```python
df = pd.read_excel('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls')
```

### Divide df into Categorical and Numerical data:

```python
cat=df[['Gender','EducationBackground','MaritalStatus','EmpDepartment','EmpJobRole',
        'BusinessTravelFrequency','EmpEducationLevel','EmpEnvironmentSatisfaction',
        'EmpJobInvolvement','EmpJobLevel','EmpJobSatisfaction','NumCompaniesWorked',
        'OverTime','EmpRelationshipSatisfaction','TrainingTimesLastYear','EmpWorkLifeBalance','Attrition']]

num=df[['Age','DistanceFromHome','EmpHourlyRate','EmpLastSalaryHikePercent','TotalWorkExperienceInYears',
        'ExperienceYearsAtThisCompany','ExperienceYearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
```

### Pie chart of Performance Rating (Target):

```python
performance_counts = df['PerformanceRating'].value_counts()
plt.pie(performance_counts, labels=performance_counts.index, autopct='%1.1f%%')
plt.show()
```

### Univariate Analysis of Categorical data:

```python
plt.figure(figsize=(10,4))
sns.countplot(data=cat,x='EducationBackground')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=cat,x='EmpDepartment')
plt.show()
```

```python
sns.countplot(data=cat,y='EmpJobRole')
plt.show()
```

```python
plt.figure(figsize=(15,10))
i=1
for x in cat.drop(['EducationBackground','EmpDepartment','EmpJobRole'],axis=1):
  plt.subplot(5,3,i)
  sns.countplot(data=cat,x=x)
  i+=1
plt.show()
```

### Univariate Analysis of Numerical data:

```python
plt.figure(figsize=(15,10))
i=1
for x in num:
  plt.subplot(3,3,i)
  sns.histplot(data=num,x=x,kde=True)
  i+=1
plt.show()
```

### Bivariate Analysis of Categorical data:

```python
plt.figure(figsize=(10,4))
sns.barplot(data=cat,x='EducationBackground',y=df['PerformanceRating'])
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=cat,x='EmpDepartment',y=df['PerformanceRating'])
plt.show()
```

```python
sns.barplot(data=cat,x=df['PerformanceRating'],y='EmpJobRole')
plt.show()
```

```python
plt.figure(figsize=(13,16))
i=1
for j in cat.drop(['EducationBackground','EmpDepartment','EmpJobRole'],axis=1):
  plt.subplot(5,3,i)
  sns.barplot(data=cat,x=j,y=df['PerformanceRating'])
  i+=1
plt.show()
```

### Bivariate Analysis of Numerical data:

```python
plt.figure(figsize=(15,10))
i=1
for j in num:
  plt.subplot(3,3,i)
  sns.barplot(data=num,x=df['PerformanceRating'],y=j)
  i+=1
plt.show()
```

### Boxplot of Numerical data:

```python
plt.figure(figsize=(12,12))
i=1
for x in num:
  plt.subplot(3,3,i)
  sns.boxplot(data=num,x=x)
  i+=1
plt.show()
```

### Heatmap correlation

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=num.corr(),annot=True,cmap='coolwarm')
plt.show()
```
